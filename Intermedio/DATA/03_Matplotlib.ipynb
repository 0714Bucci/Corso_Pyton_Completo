## Visualizzazione con MatplotLib e Saeborn
Queste librerie creano grafici e visualizzazioni intuitive per interpretare facilmente i dati.
- pip install pandas matpplotlib



```python
import matplotlib.pyplot as plt
import pandas as pd

#definizione dei dati da visualizzare
corsi = ['Python', 'java', 'javascript', 'c++', 'ruby', 'php', 'c#','go', 
        'swift', 'kotlin', 'symfony', 'django', 'flask', 'spring', 
        'angular', 'react', 'veu', 'nodejs', 'express', 'lavarel']

students = [100, 80, 90, 70, 60, 50, 40, 30, 20, 10, 5, 15, 25, 35, 45, 55, 65, 75, 85, 95]

df = pd.DataFrame({'Studenti': students, 'Corsi': corsi})
df
```

Grafico a barra (MatplotLib)

```python
plt.bar(df['Corsi'], df['Studenti'])

plt.show()
```

```python
plt.barh(df['Corsi'], df['Studenti'])

plt.show()
```

```python
plt.barh(df['Corsi'], df['Studenti'], color='yellow')
#personalizzazione del grafico
plt.title('Numero di studenti per corso')
plt.xlabel('Numero studenti')
plt.ylabel('Corsi')

plt.show()
```

### 2 Grafico a barra raggruppato(MatplotLib)
<i>Utile per: Confrontare i voti di due diverse classi (es: ClasseA, ClasseB)</i>

```python
import numpy as np
import matplotlib.pyplot as plt

corsi_predefiniti = ['Matematica', 'Fisica', 'Chimica', 'Biologia',
                     'Informatica', 'Francese','Inglese', 'Spagnolo', 
                     'Tedesco', 'Italiano']

voti_classeA = [8, 7, 9, 6, 10, 5, 7, 8, 6, 9]
voti_classeB = [6, 5, 7, 8, 9, 4, 6, 7, 5, 8] 
```

```python
#definizione delle etichette 
x = np.arange(len(corsi_predefiniti))
width = 0.35 # larghezza delle barre

fig, ax = plt.subplots() 
rects1 = ax.bar(x - width/2, voti_classeA, width, label='Classe A')
rects2 = ax.bar(x + width/2, voti_classeB, width, label='Classe B')

plt.show()
```

```python
x = np.arange(len(corsi_predefiniti))
width = 0.35 # larghezza delle barre

fig, ax = plt.subplots() 
rects1 = ax.barh(x - width/2, voti_classeA, width, label='Classe A')
rects2 = ax.barh(x + width/2, voti_classeB, width, label='Classe B')

plt.show()
```
